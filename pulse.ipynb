```python
#DataFrame of Aggregated Transaction list

path1="D:/phonepay/data/aggregated/transaction/country/india/state/"
agg_transc_list=os.listdir(path1)
agg_transc_list
col1={'State':[],'Year':[],'Quarter':[],'Transaction_Type':[],'Transaction_Count':[],'Transaction_Amount':[]}

for state in agg_transc_list:
    actual_state=path1 + "/"+state + "/"
    agg_yr=os.listdir(actual_state)
    for year in agg_yr:
        actual_yr=actual_state +"/"+ year + "/"
        agg_yr_list=os.listdir(actual_yr)
        for file in agg_yr_list:
            actual_file=actual_yr+file
            Data=open(actual_file,'r')
            D1=json.load(Data)
            for i in D1['data']['transactionData']:
                Name=i['name']
                Count=i['paymentInstruments'][0]['count']
                Amount=i['paymentInstruments'][0]['amount']
                col1['Transaction_Type'].append(Name)
                col1['Transaction_Count'].append(Count)
                col1['Transaction_Amount'].append(Amount)
                col1['State'].append(state)
                col1['Year'].append(year)
                col1['Quarter'].append(int(file.strip('.json')))

agg_trans=pd.DataFrame(col1)
agg_trans
```

```python
#DataFrame of Aggregated User List

path2 = "D:/phonepay/data/aggregated/user/country/india/state/"
agg_user_list=os.listdir(path2)
agg_user_list
col2={'State':[],'Year':[],'Quarter':[],'Brand_Name':[],'Count':[],'Percentage':[]}

for state in agg_user_list:
    actual_state=path2 + "/"+state + "/"
    agg_yr=os.listdir(actual_state)
    for year in agg_yr:
        actual_yr=actual_state +"/"+ year + "/"
        agg_yr_list=os.listdir(actual_yr)
        for file in agg_yr_list:
            actual_file=actual_yr+file
            Data=open(actual_file,'r')
            D2=json.load(Data)
            try:
                for i in D2['data']['usersByDevice']:
                    Brand_Name=i['brand']
                    Count=i['count']
                    percentage=i['percentage']
                    col2['Brand_Name'].append(Brand_Name)
                    col2['Count'].append(Count)
                    col2['Percentage'].append(percentage)
                    col2['State'].append(state)
                    col2['Year'].append(year)
                    col2['Quarter'].append(int(file.strip('.json')))
            except:
                pass

agg_user=pd.DataFrame(col2)
agg_user
```

```python
#DataFrame for MAP Transactions


path3="D:/phonepay/data/map/transaction/hover/country/india/state/"
map_transc_list=os.listdir(path3)
map_transc_list
col3={'State':[],'Year':[],'Quarter':[],'District':[],'Count':[],'Amount':[]}

for state in map_transc_list:
    actual_state=path3 + "/"+state + "/"
    map_yr=os.listdir(actual_state)
    for year in map_yr:
        actual_yr=actual_state +"/"+ year + "/"
        map_yr_list=os.listdir(actual_yr)
        for file in map_yr_list:
            actual_file=actual_yr+file
            Data=open(actual_file,'r')
            D3=json.load(Data)

            for i in D3['data']['hoverDataList']:
                district=i['name']
                Count=i['metric'][0]['count']
                Amount=i['metric'][0]['amount']
                col3['District'].append(district)
                col3['Count'].append(Count)
                col3['Amount'].append(Amount)
                col3['State'].append(state)
                col3['Year'].append(year)
                col3['Quarter'].append(int(file.strip('.json')))

map_trans=pd.DataFrame(col3)
map_trans
```

```python
#DataFrame for MAP Users

path4="D:/phonepay/data/map/user/hover/country/india/state/"
map_user_list=os.listdir(path4)
map_user_list
col4={'State':[],'Year':[],'Quarter':[],'District':[],'Registered_Users':[],'App_Opens':[]}

for state in map_user_list:
    actual_state=path4+ "/"+state + "/"
    map_yr=os.listdir(actual_state)

    for year in map_yr:
        actual_yr=actual_state +"/"+ year + "/"
        map_yr_list=os.listdir(actual_yr)

        for file in map_yr_list:
            actual_file=actual_yr+file

            Data=open(actual_file,'r')
            D4=json.load(Data)

            try:
            
                for i in D4['data']['hoverData'].items():
                        district =i[0]
                        registereduser=i[1]['registeredUsers']
                        appopens=i[1]['appOpens']
                        col4['District'].append(district)
                        col4['Registered_Users'].append(registereduser)
                        col4['App_Opens'].append(appopens)
                        col4['State'].append(state)
                        col4['Year'].append(year)
                        col4['Quarter'].append(int(file.strip('.json')))
            except:
                 pass

map_user=pd.DataFrame(col4)
map_user
```

```python
#DataFrame for Top Transactions

path5 = "D:/phonepay/data/top/transaction/country/india/state/"
top_transc_list=os.listdir(path5)
col5 = { 'State':[],'Year':[],'Quarter':[],'Pincode':[],'Transaction_Count':[],'Transaction_Amount':[]}

for state in top_transc_list:
    actual_state=path5+ "/"+state + "/"
    top_yr=os.listdir(actual_state)

    for year in top_yr:
        actual_yr=actual_state +"/"+ year + "/"
        top_yr_list=os.listdir(actual_yr)

        for file in top_yr_list:
            actual_file=actual_yr+file

            Data=open(actual_file,'r')
            D5=json.load(Data)
            
            for i in D5['data']['pincodes']:
                Name=i['entityName']
                Count=i['metric']['count']
                Amount=i['metric']['amount']
                col5['Pincode'].append(Name)
                col5['Transaction_Count'].append(Count)
                col5['Transaction_Amount'].append(Amount)
                col5['State'].append(state)
                col5['Year'].append(year)
                col5['Quarter'].append(int(file.strip('.json')))

top_trans=pd.DataFrame(col5)
top_trans
```

```python
#DataFrame for Top users

path6 = "D:/phonepay/data/top/user/country/india/state/"
top_user_list=os.listdir(path6)
col6 = { 'State':[],'Year':[],'Quarter':[],'Pincode':[],'Registered_Users':[]}

for state in top_user_list:
    actual_state=path6+ "/"+state + "/"
    top_yr=os.listdir(actual_state)

    for year in top_yr:
        actual_yr=actual_state +"/"+ year + "/"
        top_yr_list=os.listdir(actual_yr)

        for file in top_yr_list:
            actual_file=actual_yr+file

            Data=open(actual_file,'r')
            D6=json.load(Data)
            
            for i in D6['data']['pincodes']:
                Name=i['name']
                registeredusers=i['registeredUsers']
                col6['Pincode'].append(Name)
                col6['Registered_Users'].append(registeredusers)                
                col6['State'].append(state)
                col6['Year'].append(year)
                col6['Quarter'].append(int(file.strip('.json')))

top_user=pd.DataFrame(col6)
top_user
```

```python
#Connecting Mysql database


import mysql.connector as db
mydb = db.connect(  
 host="localhost",
 user="root",
 password="",
 database="phonepe"
)

mycursor = mydb.cursor(buffered=True)

mycursor.execute('create database if not exists phonepe')
```

```python
#create agg trans table

mycursor.execute("""CREATE TABLE IF NOT EXISTS  aggregate_transaction (State varchar(100),
                  Year int,
                  Quarter int,
                  Transaction_Type varchar(100),
                  Transaction_Count int, 
                  Transaction_Amount double)""")


for i, row in agg_trans.iterrows():
      sql = "INSERT INTO aggregate_transaction (State, Year, Quarter, Transaction_Type, Transaction_Count, Transaction_Amount) VALUES (%s, %s, %s, %s, %s, %s)"
      values = (row['State'], row['Year'], row['Quarter'], row['Transaction_Type'], row['Transaction_Count'], row['Transaction_Amount'])
      mycursor.execute(sql, values)
      mydb.commit()
```

```python
#create agg user table

mycursor.execute("""CREATE TABLE IF NOT EXISTS aggregate_user (State varchar(100),
                 Year int, 
                 Quarter int, 
                 Brand_Name varchar(100), 
                 Count int,
                 Percentage double)""") 

for i, row in agg_user.iterrows():
    sql = "INSERT INTO aggregate_user (State, Year, Quarter, Brand_Name, Count, Percentage) VALUES (%s, %s, %s, %s, %s, %s)"
    val = (row['State'], row['Year'], row['Quarter'], row['Brand_Name'], row['Count'], row['Percentage'])
    mycursor.execute(sql, val)
    mydb.commit()
```

```python
#create map trans table

mycursor.execute("""CREATE TABLE IF NOT EXISTS map_transaction( State varchar(100), 
                        Year int,
                        Quarter int, 
                        District varchar(100),
                        Count int, 
                        Amount double)""")

for i, row in map_trans.iterrows():
    sql = "INSERT INTO map_transaction (State, Year, Quarter, District, Count, Amount) VALUES (%s, %s, %s, %s, %s, %s)"
    val = (row['State'], row['Year'], row['Quarter'], row['District'], row['Count'], row['Amount'])
    mycursor.execute(sql, val)
    mydb.commit()
```

```python
#create map user table

mycursor.execute("""CREATE TABLE IF NOT EXISTS map_users( State varchar(100), 
                                            Year int,
                                            Quarter int, 
                                            District varchar(100), 
                                            Registered_Users int, 
                                            App_Opens int)""")

for i, row in map_user.iterrows():
    sql = "INSERT INTO map_users (State, Year, Quarter, District, Registered_Users, App_Opens) VALUES (%s, %s, %s, %s, %s, %s)"
    values = (row['State'], row['Year'], row['Quarter'], row['District'], row['Registered_Users'], row['App_Opens'])
    mycursor.execute(sql, values)
    mydb.commit()
```

```python
#create top transaction table


mycursor.execute("""CREATE TABLE IF NOT EXISTS top_transaction (
                        State varchar(100),
                        Year int,
                        Quarter int, 
                        Pincode int, 
                        Transaction_Count int,
                        Transaction_Amount double)""")

for i, row in top_trans.iterrows():
    sql = "INSERT INTO top_transaction (State, Year, Quarter, Pincode, Transaction_Count, Transaction_Amount) VALUES (%s, %s, %s, %s, %s, %s)"
    val = (row['State'], row['Year'], row['Quarter'], row['Pincode'], row['Transaction_Count'], row['Transaction_Amount'])
    mycursor.execute(sql, val)
    mydb.commit()
```
